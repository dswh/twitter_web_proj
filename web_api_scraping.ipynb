## Omicron tagged tweets

```python
query = "#omicron"

data = {
    'text': [],
    'created_at': [],
    'author_id': []
}


for tweet in tweepy.Paginator(client.search_recent_tweets, query=query, 
                              tweet_fields = ['created_at', 'author_id']).flatten(limit=1000):
    data['text'].append(tweet.text)
    data['created_at'].append(tweet.created_at)
    data['author_id'].append(tweet.author_id)
```

```python
len(data['text'])
```

```python
import pandas as pd

df = pd.DataFrame(data)
df.head()
```

```python
def extract_categories(baseurl):
    categories = {}
    r = requests.get(baseurl)
    soup = BeautifulSoup(r.content)
    categories_list = soup.find("ul", class_="nav-list").ul.find_all('li')
    for li in categories_list:
        categories.update({li.text.strip(): baseurl + li.a['href']})
    return categories
```

```python
data = extract_categories(baseurl)
```
